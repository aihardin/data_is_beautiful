```python
def rev_ratio_list(x):
    rev_ratio = zeros(len(x))
    #print x
    ratios = map(reverse_int,x)
    #print x/ratios
    #print ratios
    return ratios 
    

def reverse_int(x):
    string_x = str(x)
    #print string_x
    reverse_x = string_x[::-1]
    try:
        ratio = float(x)/int(reverse_x)
    except ZeroDivisionError:
        return 1
    except:
        print 'Error',x, reverse_x
    return ratio
```

```python
max_int = 20000
x = arange(max_int)
#f = figure()
p = subplot(111,axisbg='black')
p.spines['bottom'].set_color('white')
p.spines['left'].set_color('white')
p.tick_params(axis='x', colors='white')
p.tick_params(axis='y', colors='white')
p.get_xaxis().tick_bottom()
p.get_yaxis().tick_left()
ax = p.scatter(x,rev_ratio_list(x),s=1,c='teal',edgecolor='None')
ylim((0,5))
xlim((0,max_int))
savefig('reverse_integer_ratios.png',dpi=600,facecolor='black',bbox_inches='tight')
```
